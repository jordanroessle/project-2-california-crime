```python
#Import dependencies 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json

#for the db
from sqlalchemy import create_engine 
from config import user, password
```

```python
csv_path = "CA_Police_Dept.csv"
cali_crime = pd.read_csv(csv_path)
cali_crime 
```

```python
cali_crime = cali_crime.drop(columns=['Unnamed: 0', 'Ori'])
cali_crime 
```

```python
cali_crime.to_csv('CA_Police_Dept.csv')
cali_crime 
```

```python
#for the engine 
conn_string = f'postgresql://{user}:{password}@localhost:5432/california_crime_db'
engine = create_engine(conn_string)

#load dataframe into db 
cali_crime.to_sql(name= 'police_depts', con=engine, if exists='replace') 
```

```python

pd.read_sql_query('SELECT * FROM police_depts', con=engine).head() 
```

```python
#making the data frame to list total data per county 

county_crime = cali_crime.groupby(['County']).agg('sum')
county_crime
```

```python
county_crime.to_csv('CA_County_Crime.csv')
```
